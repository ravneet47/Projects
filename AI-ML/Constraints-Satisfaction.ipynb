# Who Killed Agatha? Dreadsbury Mansion Mystery

### Description

Someone in Dreadsbury Mansion killed Aunt Agatha. Agatha, the butler, and Charles live in
Dreadsbury Mansion, and are the only ones to live there. A killer always hates, and is no richer than
his victim. Charles hates no one that Agatha hates. Agatha hates everybody except the butler. The
butler hates everyone not richer than Aunt Agatha. The butler hates everyone whom Agatha hates.
No one hates everyone.

Who killed Agatha?

## Rules & Conclusions

1. A killer always hates, and is no richer than its victim ---> **Killer hates Agatha, Killer is no richer than Agatha**
2. Charles hates no one that Agatha hates ---> **Charles hates Butler**
3. Agatha hates everybody but the butler ---> **Agatha hates Charles & Agatha**
4. The butler hates everyone not richer than Agatha ---> **Charles & Butler**
5. The butler hates everyone whom Agatha hates ---> **Butler hates Charles & Agatha**
6. No one hates everyone ---> **Someone only hates 2 people including themselves**

### Import libraries

```python
from constraint import Problem, AllDifferentConstraint
```

### Function to solve the mystery

```python
def agatha_mystery():
    problem = Problem()
    suspects = ['Agatha', 'Butler', 'Charles', 'Killer']
    problem.addVariables(suspects, suspects)

    # Each person is a unique suspect
    problem.addConstraint(AllDifferentConstraint())

    # Adding constraint for Rule 1 is not useful since the killer can kill themselves

    # Rule 2
    def rule_2(charles, agatha):
        return charles != 'Butler'

    problem.addConstraint(rule_2, ('Charles', 'Agatha'))

    # Rule 3
    def rule_3(agatha, butler):
        return agatha != 'Charles'

    problem.addConstraint(rule_3, ('Agatha', 'Butler'))

    # Rule 4
    def rule_4(butler, agatha, charles):
        return butler != 'Killer' and butler != 'Charles'

    problem.addConstraint(rule_4, ('Butler', 'Agatha', 'Charles'))

    # Rule 5
    def rule_5(butler, agatha):
        return butler != 'Charles'

    problem.addConstraint(rule_5, ('Butler', 'Agatha'))

    # Rule 6
    def rule_6(*suspects):
        return len(set(suspects)) == 4

    problem.addConstraint(rule_6, suspects)

    solutions = problem.getSolutions()
    return solutions

solutions = agatha_mystery()
if solutions:
    print("The killer is", solutions[0]['Killer'])
else:
    print("No killer found!")
```

### Result & Conclusion

According to the constraints defined, the answer can vary.
Therefore, with these constraints, the killer is coming out to be Agatha herself.
